2. How does broadcasting work in NumPy ?
   - Broadcasting in NumPy describes how NumPy treats arrays with different   shapes during arithmetic operations [1]. Subject to certain constraints, the smaller array is "broadcast" across the larger array so that they have compatible shapes [1]. This allows for arithmetic operations between arrays of different shapes without explicitly creating multiple copies of the smaller array.

Here's a breakdown of the general rules for broadcasting:

(1).Dimension Check: If the arrays have different numbers of dimensions, the shape of the one with fewer dimensions is padded with ones on its left side.

(2).Size Check: Starting from the trailing dimensions, the sizes of the dimensions are compared.

       - If the dimensions are equal, the arrays are compatible in that dimension.
       - If one of the dimensions is 1, the other dimension is used.
       - If the dimensions are different and neither is 1, the arrays are incompatible, and a ValueError is raised.

4. Explain the use of the groupby() method in Pandas ?
   - The groupby() method in Pandas is used to group data in a DataFrame based  on one or more criteria. This operation involves some combination of splitting the object, applying a function, and combining the results [1]. It's a core operation for performing aggregate analysis on data.

- Here's a breakdown of how it works and its uses:

(1). Splitting: The groupby() method splits the DataFrame into smaller pieces   (groups) based on the unique values in the specified column(s) or other criteria. Each group is essentially a sub-DataFrame containing rows that share the same value(s) in the grouping column(s).

(2). pplying: After splitting, you can apply various functions to each group. These functions can be:

         - Aggregation: Calculating a single value for each group (e.g., mean, sum, count, min, max).
         - Transformation: Performing a group-specific computation that results in a DataFrame or Series with the same index as the original group (e.g., filling missing values within each group).
         - Filtration: Discarding or keeping groups based on a condition (e.g., keeping groups that have a certain number of rows).

(3). Combining: Finally, the results from applying the function to each group are combined back into a single DataFrame or Series.

When performing operations between arrays with incompatible shapes, NumPy automatically broadcasts the smaller array to match the shape of the larger array, subject to certain rules.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
    - NumPy's array slicing is faster than Python's list slicing primarily due to the fundamental differences in how NumPy arrays and Python lists are implemented and managed in memory:

(1). Homogeneous Data Type and Contiguous Memory:

   - NumPy Arrays: NumPy arrays store elements of the same data type in a contiguous block of memory. This means that the elements are stored next to each other in memory. When you slice a NumPy array, NumPy can efficiently calculate the memory location of the start and end of the slice and return a "view" of the original array (a reference to the same data) without copying the data.

   - Python Lists: Python lists can store elements of different data types and are not necessarily stored contiguously in memory. A Python list is essentially a list of pointers to objects scattered throughout memory. When you slice a Python list, Python needs to create a new list and copy the pointers to the objects within the slice. This copying operation is slower than creating a view.

(2). Underlying Implementation:

   - NumPy Arrays: NumPy is built on top of a C library, which allows for low-level memory management and optimized operations. NumPy's slicing operations are implemented in C, which is significantly faster than Python's interpreted code.

   - Python Lists: Python lists are implemented in Python's core, which is an interpreted language. The slicing operation for lists involves more overhead due to the need to handle potentially different data types and the dynamic nature of the list.

(3). No Type Checking Overhead:

  - NumPy Arrays: Since NumPy arrays are homogeneous (all elements have the same data type), NumPy doesn't need to perform type checking for each element during slicing.

  - Python Lists: Python lists can contain elements of different types, so Python needs to perform type checking during slicing to ensure compatibility and handle different object types.

Ques 11.  Perform matrix multiplication using NumPy?

```python
import numpy as np

# Create two NumPy matrices (2D arrays)
matrix1 = np.array([[1, 2],
                    [3, 4]])

matrix2 = np.array([[5, 6],
                    [7, 8]])

# Perform matrix multiplication
# Method 1: Using the @ operator (preferred in recent Python/NumPy versions)
result_matrix_1 = matrix1 @ matrix2

# Method 2: Using np.dot()
result_matrix_2 = np.dot(matrix1, matrix2)

# Method 3: Using the .dot() method of a NumPy array
result_matrix_3 = matrix1.dot(matrix2)


# Print the original matrices and the results
print("Matrix 1:")
print(matrix1)

print("\nMatrix 2:")
print(matrix2)

print("\nMatrix multiplication result (using @ operator):")
print(result_matrix_1)

print("\nMatrix multiplication result (using np.dot()):")
print(result_matrix_2)

print("\nMatrix multiplication result (using .dot() method):")
print(result_matrix_3)
```

Ques 13. Create a 3D scatter plot using Plotly

```python
import plotly.graph_objects as go
import numpy as np

# Generate some sample 3D data
np.random.seed(42) # for reproducibility
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)
z = np.random.rand(num_points)
sizes = np.random.rand(num_points) * 30 # Optional: for bubble chart effect
colors = np.random.rand(num_points) # Optional: for coloring points


# Create the 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers', # Use 'markers' to create a scatter plot
    marker=dict(
        size=sizes, # Use the sizes array for varying marker sizes
        color=colors, # Use the colors array for varying marker colors
        colorscale='Viridis', # Choose a colorscale (optional)
        opacity=0.8
    )
)])

# Update layout for title and axis labels
fig.update_layout(
    title='Simple 3D Scatter Plot',
    scene = dict(
        xaxis_title='X-axis',
        yaxis_title='Y-axis',
        zaxis_title='Z-axis'
    ),
    margin=dict(l=0, r=0, b=0, t=40) # Adjust margins (optional)
)

# Display the plot
fig.show()
```
